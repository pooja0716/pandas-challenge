```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Find count of unique items
unique_items=len(purchase_data_df["Item Name"].unique())

#Find count of purchases
no_of_items=(purchase_data_df["Item ID"].count())


#Find total revenue
Sum_Price=purchase_data_df["Price"].sum()

#Find average price
Average_Price=round(Sum_Price/no_of_items,2)

#print(Average_Price)
#items_count=purchase_data_df()

#create dataframe
purchasing_analysis_df=pd.DataFrame({'Number of Unique Items': [unique_items],
                                 "Average Price": [Average_Price],
                                 "Number of Purchases": [no_of_items],
                                 "Total Revenue": [Sum_Price]})


purchasing_analysis_df["Total Revenue"] =purchasing_analysis_df["Total Revenue"].map("${:,.2f}".format)
purchasing_analysis_df["Average Price"] =purchasing_analysis_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_df
```

```python
#Retrieve the Item ID, Item Name, and Item Price columns
most_popular_index_loc=purchase_data_df.loc[:,["Item ID","Item Name"]]

#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
most_popular_group_df=purchase_data_df.groupby(['Item ID',"Item Name"])

most_popular_total_purchase = most_popular_group_df["Price"].sum()

most_popluar_count_purchase= most_popular_group_df["Purchase ID"].count()

most_popular_item_price=most_popular_total_purchase/most_popluar_count_purchase

#Create a summary data frame to hold the result
most_popular_ana_df=pd.DataFrame({"Purchase Count":most_popluar_count_purchase,
                                  "Item Price":most_popular_item_price,
                                  "Total Purchase Value":most_popular_total_purchase})
#Sort the purchase count column in descending order
most_popular_ana_df=most_popular_ana_df.sort_values("Purchase Count", ascending=False)

#Optional: give the displayed data cleaner formatting
most_popular_ana_df["Item Price"]=most_popular_item_price.map("${:,.2f}".format)
most_popular_ana_df["Total Purchase Value"]=most_popular_total_purchase.map("${:,.2f}".format)

#Display a preview of the summary data frame
most_popular_ana_df.head(5)
```
